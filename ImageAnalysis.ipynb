```python
import imageio
from skimage import exposure
from skimage import img_as_ubyte
import numpy as np
from scipy import ndimage
import os
import matplotlib.pyplot as plt
from skimage.morphology import disk
from skimage import filters

from skimage import restoration


filename = os.path.join('/Users/jmukherj/Documents/research/python', 'DANCE.JPG')
camera = imageio.imread(filename)
plt.imshow(camera)

blurred_f = ndimage.gaussian_filter(camera, 0)
#increase the weight of edges by adding an approximation of the Laplacian:
plt.imshow(blurred_f)
 
filter_blurred_f = ndimage.gaussian_filter(blurred_f, 1)
plt.imshow(filter_blurred_f)

alpha = 30
sharpened = blurred_f + alpha * (blurred_f - filter_blurred_f)
plt.imshow(sharpened)
```
